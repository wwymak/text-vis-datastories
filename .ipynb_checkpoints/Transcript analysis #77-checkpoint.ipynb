```python
parsedTextCleaned = [text for text in parsedText if len(text) > 0]
```

```python
print(parsedTextCleaned[0: 10])
```

```python
ebSpeech = [text for text in parsedTextCleaned if (text.startswith("Enrico Bertini:")  or text.startswith("EB:"))]
```

```python
len(ebSpeech)
```

```python
msSpeech = [text for text in parsedTextCleaned if (text.startswith("Moritz Stefaner:")  or text.startswith("MS:"))]
```

```python
mdSpeech = [text for text in parsedTextCleaned if (text.startswith("Matt Daniels:")  or text.startswith("MD:"))]
```

```python
posEB = [nltk.pos_tag(nltk.word_tokenize(text)) for text in ebSpeech]
posMS = [nltk.pos_tag(nltk.word_tokenize(text)) for text in msSpeech]
posMD = [nltk.pos_tag(nltk.word_tokenize(text)) for text in mdSpeech]
```

```python
posEB
```

```python
ebWords = [len(x) for x in posEB]
```

```python
ebWords
```

```python
blobsEB = [TextBlob(text) for text in ebSpeech]
blobsMS = [TextBlob(text)  for text in msSpeech]
blobsMD = [TextBlob(text)  for text in mdSpeech]
```

```python
sentencesEB = [textblob.sentences for textblob in blobsEB]
sentencesMS = [textblob.sentences for textblob in blobsMS]
sentencesMD = [textblob.sentences for textblob in blobsMD]
```

```python
sentimentSentenceEB = [[a.sentiment for a in sentenceArr] for sentenceArr in sentencesEB]
sentimentSentenceMS = [[a.sentiment for a in sentenceArr] for sentenceArr in sentencesMS]
sentimentSentenceMD = [[a.sentiment for a in sentenceArr] for sentenceArr in sentencesMD]
```

```python
sentimentEB = [textblob.sentiment for textblob in blobsEB]
sentimentMS = [textblob.sentiment for textblob in blobsMS]
sentimentMD = [textblob.sentiment for textblob in blobsMD]
```

```python
print(sentencesEB[3], sentimentSentenceEB[3])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tf = TfidfVectorizer(analyzer='word', ngram_range=(1,3), min_df = 0, stop_words = 'english')
```

```python
import json
ebSentences = json.dumps([[str(txt) for txt in sArr] for sArr in sentencesEB])
msSentences = json.dumps([[str(txt) for txt in sArr] for sArr in sentencesMS])
mdSentences = json.dumps([[str(txt) for txt in sArr] for sArr in sentencesMD])
```

```python
with open('ebSentences.json', 'w') as outfile:
    json.dump(ebSentences, outfile)
```

```python
ebSentimentJson = json.dumps([[{"polarity": x.polarity, "subjectivity": x.subjectivity} for x in sArr] for sArr in sentencesEB])
mdSentimentJson = json.dumps([[{"polarity": x.polarity, "subjectivity": x.subjectivity} for x in sArr] for sArr in sentencesMD])
msSentimentJson = json.dumps([[{"polarity": x.polarity, "subjectivity": x.subjectivity} for x in sArr] for sArr in sentencesMS])
```

```python
with open('ebSentiment.json', 'w') as outfile:
    json.dump(ebSentimentJson, outfile)
```

```python
with open('msSentences.json', 'w') as outfile:
    json.dump(msSentences, outfile)
with open('mdSentences.json', 'w') as outfile:
    json.dump(mdSentences, outfile)
```

```python
with open('msSentiment.json', 'w') as outfile:
    json.dump(msSentimentJson, outfile)
with open('mdSentiment.json', 'w') as outfile:
    json.dump(mdSentimentJson, outfile)
```
